```python
import requests

import pandas as pd
url = "https://www.elasslihitech.com/pc-portable/"
response = requests.get(url)
print(response.status_code)
print(response.text[:1000])

```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text,"html.parser")
```

```python
titles = soup.find_all("h2",class_="woocommerce-loop-product__title")
prices  = soup.find_all("span",class_="woocommerce-Price-amount amount")
```

```python
import pandas as pd
product_data =[]
for t, p in zip(titles, prices):
    product_data.append({"Title": t.get_text(strip=True),
    "Price": p.get_text(strip=True)})
df= pd.DataFrame(product_data)
print(df.head())
print(len(df))
```

```python
all_products =[]
page =1
while page<5:
    print("Scraping page", page)
    url = url.format(page)
    response = requests.get(url )
    if response.status_code != 200:
        print("No more pages.")
        break
    soup= BeautifulSoup(response.text,"html.parser")
    titles= soup.find_all("h2", class_="woocommerce-loop-product__title")
    prices = soup.find_all("span", class_="woocommerce-Price-amount")
    if not titles or not prices:
        break
    for t,p in zip(titles, prices):
        all_products.append({"Title": t.get_text(strip=True),
        "Price": p.get_text(strip=True)})
    page += 1
df = pd.DataFrame(all_products)
df.to_csv("admininfo_all_products.csv", index=False, encoding="utf-8")
print("Saved admininfo_all_products.csv successfully!")
```

```python
df["Price"] = (
    df["Price"]
    .str.replace("DA", "", regex=False)  
    .str.replace(" ", "", regex=False)  
    .astype(float)                       
)

mean_price = df["Price"].mean()
print("Average price:", mean_price)
```
