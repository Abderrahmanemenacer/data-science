```python
df.fillna({"Phone Number":"unkonwn"},inplace=True)
df.fillna({"Postcode":"unkonwn"},inplace=True)
df.fillna({"Longitude":df["Longitude"].mean()},inplace=True)
df.fillna({"Latitude":df["Latitude"].mean()},inplace=True)
df.fillna({"Street Address":"unkonwn"},inplace=True)
df.fillna({"City":"unkonwn"},inplace=True)
```

```python
df['Timezone_offset'] = df['Timezone'].str.extract(r'([+-]\d+)').astype(float)
df['Timezone_offset'].std()
```
